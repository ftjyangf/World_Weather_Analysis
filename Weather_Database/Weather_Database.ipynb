```python
import pandas as pd
import scipy.stats as sci
import numpy as np
from citipy import citipy
```

```python
lats = np.random.randint(-90,89,2000) + np.random.random(2000)
longs = np.random.uniform(-180,180,2000)
coordinates = zip(lats,longs)
coordinates =list(coordinates)
```

```python
cities =[]
for each in coordinates:
    city = citipy.nearest_city(each[0],each[1]).city_name
    if not city in cities:
        cities.append(city)
        

        
    
```

```python
import requests
from config import weather_api_key
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# Latitude and longitude
# Maximum temperature
# Percent humidity
# Percent cloudiness
# Wind speed
# Weather description (for example, clouds, fog, light rain, clear sky)
summer_data_df = pd.DataFrame()
```

```python
weather_data = list()
set_count = 1
record_count = 1
print("Start to retrieve data")
for i, city in enumerate(cities):
    if i % 50 == 0 and i >= 50:
        set_count +=1
        record_count = 1 
    print('-----------------------')
    print(f"processing record {record_count} of set {set_count}")
    print('-----------------------')
    try: 
        city_url = url + "&q=" + city.replace(" ", "+")
        city_weather = requests.get(city_url).json()
        country = city_weather['sys']['country']
        lats = city_weather['coord']['lat']
        lon =  city_weather['coord']['lon']
        max_t =  city_weather['main']["temp_max"]
        humdility =  city_weather['main']['humidity']
        cloudiness =  city_weather['clouds']['all']
        wind_speed =  city_weather['wind']['speed']
        desc =  city_weather['weather'][0]['description']
        
        record_count += 1
     
        weather_data.append({
                            "City":city.title(),
                            "Country":country,
                            "Lat": lats,
                            "Lon": lon,
                            "Max Temp" : max_t,
                            "Humdiity" : humdility,                  
                            "Cloudiness" :  cloudiness,
                            "Wind Speed" :  wind_speed,
                            "Weather description" :  desc}
                             )
       
        
    except:
        print("------------------------")
        print(f"{city} has no weather data")
        print("------------------------")
print("Data retrieving finished")
```
