```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
lats = np.random.uniform(-90,90,1500).round(2)
lngs = np.random.uniform(-180,180,1500).round(2)
```

```python
lats_lngs = zip(lats,lngs)
coordinates = list(lats_lngs)
```

```python
from citipy import citipy
```

```python
for each in coordinates:
    print(citipy.nearest_city(each[0],each[1]).city_name,
          citipy.nearest_city(each[0],each[1]).country_code)
```

```python
cities = []
for each in coordinates:
    city = citipy.nearest_city(each[0],each[1]).city_name
    if city not in cities:
        cities.append(city)
    
            
len(cities)
```

```python
i = 0
weather_cities = []
while i < 3:
    city_url = url + "&q=" +cities[i] 
    city = requests.get(city_url)
    weather_cities.append(city.json())
    i = i +1
```

```python
# City, country, and date
# Latitude and longitude
# Maximum temperature
# Humidity
# cloudiness
# Wind speed
from datetime import datetime
```

```python
#city :city.json['name'],country: ['sys']['country'],#date['dt']
#lati and longi [coord]['lon']/['lat']
#max temp = "temp_max"
#humaidity ['main']['humidity']
#cloud ["cloud"]['all']
#wind speed ['wind']['speed']
print("Begining Data Retrieve  ")
print("------------------------")
name = []
country = []
date = []
lat = []
lng = []
Max_t = []
humidity = []
cloudiness = []
wind_s = []
i = 0
record_count = 1
set_count = 1


for i, city in enumerate(cities):
        if (i % 50 == 0 and i >= 50):
            set_count += 1
            record_count = 1
        print(f"processing Record {record_count} of set {set_count} | {city}")
        try:
            city_url = url + "&q=" + city
            wcities = requests.get(city_url)
            name.append(wcities.json()['name'])
            country.append(wcities.json()['sys']['country'])
            date.append(datetime.utcfromtimestamp(wcities.json()['dt']).strftime('%Y-%m-%d %H:%M:%S'))
            lat.append(wcities.json()['coord']['lat'])
            lng.append(wcities.json()['coord']['lon'])
            Max_t.append(wcities.json()['main']['temp_max'])
            humidity.append(wcities.json()['main']['humidity'])
            cloudiness.append(wcities.json()['clouds']['all'])
            wind_s.append(wcities.json()['wind']['speed'])
            record_count += 1

        except:
            print(city + " have no weather data,skipping...")
            pass
print('----------------------------')
print('Data Retrieval Complete')
print('----------------------------')
    

```

```python
cities_weather_df = pd.DataFrame({
         "name": name,
         "country" : country,
         "Date" : date,
         "Lat" : lat,
         "lng" : lng,
         "Max Temp":Max_t,
         "Humidity":humidity,
         "Cloudiness":cloudiness,
         "wind_speed":wind_s
})
```

```python
# output data
output_data_file = "weather_data/cities.csv"
#export 
cities_weather_df.to_csv(output_data_file,index_label ="City_ID")
```
