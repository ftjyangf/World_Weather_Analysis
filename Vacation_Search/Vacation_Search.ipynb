## Deliverable 2. Create a Customer Travel Destinations Map.

```python
# Dependencies and Setup
import pandas as pd
import requests
import gmaps

# Import API key
from config import g_key

# Configure gmaps API key
gmaps.configure(api_key=g_key)
```

```python
# 1. Import the WeatherPy_database.csv file. 
city_data_df = pd.read_csv("../Weather_Database/WeatherPy_database.csv")
city_data_df.head()
```

```python
# 2. Prompt the user to enter minimum and maximum temperature criteria 
try:
    #min_temp = float(input("Please type the minimum city temperature"))
    #In first deliverable min_temp wasn't asked to be in city_data_df 
    max_temp = float(input("Please type the maxium city temperature"))
except:
    print("Please type only number")
```

```python
# 3. Filter the city_data_df DataFrame using the input statements to create a new DataFrame using the loc method.
filtered_data_df = city_data_df.loc[(city_data_df['Max Temp'] < max_temp)]
```

```python
# 4a. Determine if there are any empty rows.

hotel_df =filtered_data_df[filtered_data_df['Country'].notnull()]
```

```python
# 4b. Drop any empty rows and create a new DataFrame that doesn’t have empty rows.
hotel_df
# In 5a it didn't ask for weather description, but later in gmaps this info was required so hotel_des_df need to be added to
# add aweather description to hotel_df dataframe
hotel_des_df = hotel_df
```

```python
# 5a. Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = hotel_df[['City','Country','Max Temp','Lat','Lon']].copy()

# 5b. Create a new column "Hotel Name"
hotel_df["Hotel Name"] = ""
hotel_df.reset_index(drop = True,inplace = True)
hotel_df
```

```python
# 6a. Set parameters to search for hotels with 5000 meters.
hotel_json = list()
gmaps.configure(api_key = g_key)
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
radius = 5000
params = {
        "radius": 5000,
        "type": "lodging",
        "key": g_key,
        "location":""
    }
# 6b. Iterate through the hotel DataFrame.
for i,row in enumerate(hotel_df.iterrows()):
    try:
        precentage = (i/len(hotel_df.index)) * 100
        print('---------------------------------------')
        print(f"Retrieved {precentage} percent of data")
# 6c. Get latitude and longitude from DataFrame
        location = str(row[1]['Lat'])+","+str(row[1]['Lon'])

#6d. Set up the base URL for the Google Directions API to get JSON data.
        params['location'] = location
    
# 6e. Make request and retrieve the JSON data from the search. 
        hotel_df1 = requests.get(base_url,params = params).json()
        hotel_df.loc[(i,'Hotel Name')] = hotel_df1['results'][0]['name']
  
    
    except:
        print(f"{row[1]['City']} has no data")
    
print('---------------------------------------')
print('Completed')   
# 6f. Get the first hotel from the results and store the name, if a hotel isn't found skip the city.
```

```python
# 7. Drop the rows where there is no Hotel Name.
clean_hotel_df = hotel_df.loc[hotel_df['Hotel Name'].eq("") == False]

#get weather description
clean_hotel_df = pd.merge(clean_hotel_df,hotel_des_df[['City','Weather description']],how="inner")
```

```python
# 8a. Create the output File (CSV)
output_data_file = "WeatherPy_vacation.csv"
# 8b. Export the City_Data into a csv
clean_hotel_df.to_csv(output_data_file)

```

```python
# 9. Using the template add city name, the country code, the weather description and maximum temperature for the city.
info_box_template = """
<dl>
<dt>City</dt>
<dd>{City}</dd>
<dt>Country</dt>
<dd>{Country}</dd>
<dt>Weather description</dt>
<dd>{Weather description}</dd>
<dt>Max Temp</dt>
<dd>{Max Temp} °F</dd>
</dl>
"""

# 10a. Get the data from each row and add it to the formatting template and store the data in a list.
hotel =list()
for i,city in enumerate(clean_hotel_df.iterrows()):
    City = city[1]['City']
    Country = city[1]['Country']
    Weather = city[1]['Weather description']
    Max_Temp = city[1]['Max Temp']
    
    hotel.append({
        "City":City,
        "Country":Country,
        "Weather description":Weather,
        "Max Temp":Max_Temp
    }) 
# hotel_info = [info_box_template.format(**row) for index, row in clean_hotel_df.iterrows()]
hotel_info =[info_box_template.format(**row) for row in hotel]
# 10b. Get the latitude and longitude from each row and store in a new DataFrame.

locations = clean_hotel_df[["City","Lat", "Lon"]]
```

```python
# 11a. Add a marker layer for each city to the map.
marker_locations =list()
for i in locations.iterrows():
    locat = (i[1]['Lat'],i[1]['Lon'])
    marker_locations.append(locat)
      
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)


# 11b. Display the figure
heat_layer = gmaps.heatmap_layer(hotel_df[['Lat','Lon']], weights=hotel_df['Max Temp'],dissipating=False,
             max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(marker_locations, info_box_content=hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
